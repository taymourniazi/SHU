```python
import numpy as np
from sklearn.linear_model import LinearRegression
import pandas as pd
import matplotlib.pyplot as plt
```

```python
#read data
dataset=pd.read_csv('sales.csv')
```

```python
#creating variable seperately
income=dataset['income']
credit=dataset['credit']
```

```python
#reshaping
income=income.reshape(-1,1)
credit=credit.reshape(-1,1)
```

```python
#fitting linear regression
regressor=LinearRegression()
regressor.fit(income, credit)
```

```python
#plotting scatterplot and good fit line
plt.scatter(income, credit, color ='red')
plt.plot(income, regressor.predict(income), color='black')
plt.title('scatterplot for sales data')
plt.xlabel('Income')
plt.ylabel('Credit')
plt.show()
```
