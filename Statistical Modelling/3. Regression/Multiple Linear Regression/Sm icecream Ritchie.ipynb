```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.linear_model import LinearRegression
import statsmodels.formula.api as smf
```

```python
#read data
df = pd.read_csv('icecream.csv', index_col=0)
df.head(n=10)
```

```python
df.shape
```

```python
#plotting the Icecream data distribution
sns.set()
sns.pairplot(df, x_vars=['X1', 'X2', 'X3'], y_vars='Y', size=4, aspect=0.7)
```

```python
#plotting linear regression through seaborn
sns.pairplot(df, x_vars=['X1', 'X2', 'X3'], y_vars='Y', size=4,
aspect=0.7, kind='reg')
```

```python
#fitting linear egresion model through STATSMODEL
lm = smf.ols(formula='Y ~ X1 + X2 + X3', data=df).fit()
```

```python
lm.summary()
```

```python
#fitting linear regresion model through Scikit learn

#creating array for th three predictor
x = ['X1', 'X2', 'X3']
x = df[x]
y = df.Y
regressor = LinearRegression()
regressor.fit(x, y)

print (regressor.intercept_)
list(zip(x, regressor.coef_))


```
