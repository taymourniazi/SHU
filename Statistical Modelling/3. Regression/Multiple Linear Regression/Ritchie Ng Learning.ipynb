```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.linear_model import LinearRegression
import statsmodels.formula.api as smf
```

```python
#read data
df = pd.read_csv('Advertising.csv', index_col=0)
df.head(n=10)
```

```python
df.shape
```

```python
#plotting the advertising data distribution
sns.set()
sns.pairplot(df, x_vars=['TV', 'radio', 'newspaper'], y_vars='sales', size=4)
```

```python
#plotting linear regression through seaborn
sns.pairplot(df, x_vars=['TV','radio','newspaper'], y_vars='sales', size=7,
aspect=0.7, kind='reg')
```

```python
#fitting linear egresion model through STATSMODEL
lm = smf.ols(formula='sales ~ TV + radio + newspaper', data=df).fit()
```

```python
lm.summary()
```

```python
#fitting linear regresion model through Scikit learn

#creating array for th three predictor
x = ['TV', 'radio', 'newspaper']
x = df[x]
y = df.sales
regressor = LinearRegression()
regressor.fit(x, y)

print (regressor.intercept_, regressor.coef_)
list(zip(x, regressor.coef_))
```
